```python
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
os.getcwd()
os.chdir('/Users/nitheshsrb/Documents/VSCode/Git/Fundamentals')
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
summary = df['HomePlanet'].value_counts().reset_index()
summary.columns = ['HomePlanet','counts']
summary['percentage'] = summary['counts']/summary['counts'].sum() * 100
summary['percentage'] = (round(summary['percentage']).astype('str') + '%').astype('str')

#fig, axes = plt.subplots(1,1,figsize = (10,8))
sns.barplot(data = summary, x = 'HomePlanet', y = 'counts')
for idx, (percentage,val) in enumerate(zip(summary['percentage'],summary['counts'])):
    plt.annotate(text = percentage
                 , xytext = (2,2)
                 , xy= (idx,val)
                 ,textcoords= 'offset points'
                 , ha = 'center'
                 )
```
